```python
import os

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from matplotlib.pyplot import annotate


df = pd.DataFrame(columns=['Hour of ST data', 'BLEU', 'Method'])

df['Hours of ST data'] = [1, 10] * 4

df['BLEU'] = [4.3, 17.48] + [11.66, 18.62] + [14.62, 20.38] + [15.32, 20.79]
df['Method'] = ['none'] * 2 + ['sentence'] * 2 + ['ctc'] * 2 + ['token'] * 2

sns_plot = sns.catplot(data=df, x='Hours of ST data', y='BLEU', hue='Method', kind='bar')
for c in sns_plot.ax.containers:
    sns_plot.ax.bar_label(c, fmt='%.2f', size=8)
sns_plot.figure.savefig('figures/pretrain_finetune.pdf', dpi=100)
```

```python
from matplotlib.pyplot import annotate


df = pd.DataFrame(columns=['Hour of ASR data', 'BLEU', 'Method'])

df['Hours of ASR data'] = [100, 400] * 2

df['BLEU'] = [18.26, 20.38] + [20.13, 20.79]
df['Method'] = ['ctc'] * 2 + ['token'] * 2 

sns_plot = sns.catplot(data=df, x='Hours of ASR data', y='BLEU', hue='Method', kind='bar', aspect=0.7)
for c in sns_plot.ax.containers:
    sns_plot.ax.bar_label(c, fmt='%.2f', size=8)
sns_plot.figure.savefig('figures/pretrain_finetune_100h.pdf', dpi=100)
```
