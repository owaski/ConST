```python
import os

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from matplotlib.pyplot import annotate
from numpy import size


df = pd.DataFrame(columns=['Hour of ASR data', 'BLEU', 'Method'])

df['Hours of ASR data'] = [10, 40, 400] * 4

df['BLEU'] = [4.3, 4.3, 4.3] + [7.22, 7.50, 2.41] + [1.31, 3.11, 6.31] + [10.7, 11.01, 11.26]
df['Method'] = ['none'] * 3 + ['sentence'] * 3 + ['ctc'] * 3 + ['token'] * 3 

sns_plot = sns.catplot(data=df, x='Hours of ASR data', y='BLEU', hue='Method', kind='bar')
for c in sns_plot.ax.containers:
    sns_plot.ax.bar_label(c, fmt='%.2f', size=8)
sns_plot.figure.savefig('figures/joint_train.pdf', dpi=100)
```
